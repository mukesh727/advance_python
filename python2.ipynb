```python
#1. Parking Lot Management: Track vehicle entry/exit, available spots, and compute parking fees.
import time

class ParkingLot:
    def __init__(self):
        self.vehicles = {}

    def vehicle_entry(self, number):
        entry_time = time.time()
        self.vehicles[number] = entry_time
        print("Vehicle entered:", number)

    def vehicle_exit(self, number):
        if number not in self.vehicles:
            print("Vehicle not found")
            return

        exit_time = time.time()
        entry_time = self.vehicles[number]
        hours = (exit_time - entry_time) / 3600
        fee = round(hours * 20, 2)

        print("Vehicle exited:", number)
        print("Parking fee:", fee)

        del self.vehicles[number]

parking = ParkingLot()
parking.vehicle_entry("OD012345")
parking.vehicle_exit("OD0247593")
```

```python
#2. To-Do List Manager with Categories: Add tasks, set deadlines, assign priority, and filter by project or tag.

class Task:
    def __init__(self, title, deadline, priority, category):
        self.title = title
        self.deadline = deadline
        self.priority = priority
        self.category = category

class TodoManager:
    def __init__(self):
        self.tasks = []

    def add_task(self, title, deadline, priority, category):
        task = Task(title, deadline, priority, category)
        self.tasks.append(task)

    def show_tasks(self):
        for t in self.tasks:
            print(t.title, "|", t.deadline, "|", t.priority, "|", t.category)

    def filter_by_category(self, category):
        for t in self.tasks:
            if t.category == category:
                print(t.title)

todo = TodoManager()
todo.add_task("Finish project", "2026-03-20", "High", "College")
todo.add_task("Buy groceries", "2026-03-11", "Low", "Personal")

todo.show_tasks()
```

```python
#3. Billing System for Retail Store: Scan products, apply discounts, generate bills, and record transactions.

class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

class BillingSystem:
    def __init__(self):
        self.cart = []

    def add_product(self, product):
        self.cart.append(product)

    def generate_bill(self, discount=0):
        total = sum(p.price for p in self.cart)
        total = total - (total * discount / 100)

        print("---- BILL ----")
        for p in self.cart:
            print(p.name, p.price)

        print("Total:", total)

p1 = Product("Milk", 50)
p2 = Product("Bread", 30)

bill = BillingSystem()
bill.add_product(p1)
bill.add_product(p2)

bill.generate_bill(10)
```

```python
#4. Movie Ticket Booking System: Check showtimes, book tickets, select seats, and print confirmation.
class Movie:
    def __init__(self, name, seats):
        self.name = name
        self.seats = seats

    def book_seat(self):
        if self.seats > 0:
            self.seats -= 1
            print("Ticket booked")
        else:
            print("House full")

movie = Movie("Avengers", 5)

movie.book_seat()
movie.book_seat()

print("Seats left:", movie.seats)
```

```python
#5. College Result Management: Teachers enter marks, calculate GPA ~

class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def calculate_gpa(self):
        avg = sum(self.marks) / len(self.marks)
        gpa = avg / 10
        return round(gpa, 2)

s1 = Student("Rahul", [80, 75, 90, 85])

print("Student:", s1.name)
print("GPA:", s1.calculate_gpa())
```

```python
#1. Password Manager: Securely store, add, edit, and retrieve user passwords using encryption.

import base64

class PasswordManager:
    def __init__(self):
        self.passwords = {}

    def add_password(self, site, password):
        encoded = base64.b64encode(password.encode()).decode()
        self.passwords[site] = encoded

    def get_password(self, site):
        encoded = self.passwords.get(site)
        if encoded:
            return base64.b64decode(encoded).decode()

pm = PasswordManager()

pm.add_password("gmail", "mypassword123")
print(pm.get_password("gmail"))
```

```python
#2. Freelancer Marketplace Simulation: Register freelancers and clients, assign projects, and process payments.

class Freelancer:
    def __init__(self, name):
        self.name = name

class Client:
    def __init__(self, name):
        self.name = name

class Project:
    def __init__(self, title, payment):
        self.title = title
        self.payment = payment

f = Freelancer("Amit")
c = Client("TechCorp")

project = Project("Website Design", 500)

print(f.name, "assigned to", project.title)
print("Client:", c.name)
print("Payment:", project.payment)
```

```python
#3. Warehouse Automation System: Track goods movement, generate inventory reports, and forecast demand.

class Warehouse:
    def __init__(self):
        self.inventory = {}

    def add_item(self, item, quantity):
        if item in self.inventory:
            self.inventory[item] += quantity
        else:
            self.inventory[item] = quantity

    def remove_item(self, item, quantity):
        if item in self.inventory:
            self.inventory[item] -= quantity

    def report(self):
        print("Inventory Report")
        for item, qty in self.inventory.items():
            print(item, qty)

w = Warehouse()
w.add_item("Laptop", 10)
w.add_item("Mouse", 50)
w.remove_item("Mouse", 5)

w.report()
```
