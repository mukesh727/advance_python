```python
rooms = {}
total_rooms = 5
available_rooms = list(range(1, 1 + total_rooms))

while True:
    choice = input("Do you want  a room ").lower()
    if choice != "yes":
        break

    if not available_rooms:
        print("Sorry, no rooms available.")
        break

    name = input("Enter name: ")
    members = int(input("Enter number of members: "))


    room_number = available_rooms.pop(0)
    rooms[room_number] = {"Name": name, "Members": members}

    print(f"Room {room_number} allotted to {name} for {members} members.")


print("\nAll Room Allotments:")
for room, info in rooms.items():
    print(f"Room {room}: {info['Name']} ({info['Members']} members)")
```

```python
#List based transaction system
balance = 0
transactions = []

while True:
    print("\n baning system ")
    print("1. Deposit")
    print("2. Withdraw")
    print("3. Check Balance")
    print("4. Transaction History")
    print("5. Exit")

    try:
        choice = int(input("Enter the choice: "))
    except ValueError:
        print("Invalid input")
        continue

    if choice == 1:
        try:
            amt = int(input("Enter deposit amount: "))
            if amt <= 0:
                print("Amount more than 0.")
            else:
                balance += amt
                transactions.append("Deposit: " + str(amt))
                print("Amount credited.")
        except ValueError:
            print("Invalid amount!")

    elif choice == 2:
        try:
            amt = int(input("Enter withdraw amount: "))
            if amt <= 0:
                print("Amount more then 0.")
            elif balance < amt:
                print("Insufficient balance!.")
            else:
                balance -= amt
                transactions.append("Withdraw: " + str(amt))
                print("Amount withdrawn.")
        except ValueError:
            print("Invalid amount!")

    elif choice == 3:
        print("Available Balance:", balance)

    elif choice == 4:
        print("\nTransaction History:")
        if not transactions:
            print("No transaction available.")
        else:
            for t in transactions:
                print(t)

    elif choice == 5:
        print("Thank you.")
        break

    else:
        print("Invalid choice! Please try again.")
```
